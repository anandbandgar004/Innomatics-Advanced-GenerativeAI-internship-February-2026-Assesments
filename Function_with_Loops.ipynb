Problem 1: Mobile Recharge Validation System

```python

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:  # Retry loop
        amount = int(input("Enter recharge amount: ‚Çπ"))

        # Minimum amount check
        if amount < 50:
            print("‚ùå Recharge amount must be at least ‚Çπ50. Try again.\n")
            continue

        # Valid plan check
        if amount in valid_plans:
            print("‚úÖ Recharge Successful with plan ‚Çπ", amount)
            break
        else:
            print("‚ùå Invalid plan. Valid plans are:", valid_plans)
            print("Please retry.\n")


validate_recharge()
```

Problem 2: Inventory Reorder Alert System

```python

def check_inventory(products):

    print("\nInventory Status:\n")

    # Loop through dictionary
    for product, stock in products.items():

        if stock < 15:
            print(f"üî¥ {product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"üü¢ {product}: Stock OK (Stock = {stock})")


# Sample product dictionary
inventory_data = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 18,
    "Salt": 8
}


check_inventory(inventory_data)
```

Problem 3: Student Result Processing System

```python

def process_result(marks_list):

    total = 0

    # Loop to calculate total marks
    for marks in marks_list:
        total += marks

    # Calculate average
    average = total / len(marks_list)

    print(f"\nAverage Marks = {average:.2f}")

    # Result decision
    if average >= 50:
        return "‚úÖ Pass"
    else:
        return "‚ùå Fail"



marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    m = int(input(f"Enter marks for subject {i+1}: "))
    marks.append(m)


result = process_result(marks)
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python

def calculate_fare(distance, peak):

    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # Peak hour charge
    if peak == "yes":
        fare += fare * 0.25  # 25% extra

    return fare


# Retry loop
while True:

    distance = float(input("\nEnter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)

    print(f"üöñ Total Fare = ‚Çπ{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ").lower()

    if retry != "yes":
        print("Thank you for using Fare Estimator.")
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python

def check_attendance(attendance_list):

    present_days = 0

    # Loop to count "P"
    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1

    total_days = len(attendance_list)

    percentage = (present_days / total_days) * 100

    print(f"\nAttendance Percentage = {percentage:.2f}%")

    if percentage >= 75:
        return "‚úÖ Eligible"
    else:
        return "‚ùå Not Eligible"



attendance = []

days = int(input("Enter number of working days: "))

for i in range(days):
    status = input(f"Day {i+1} (P/A): ")
    attendance.append(status)


eligibility = check_attendance(attendance)
print("Eligibility Status:", eligibility)
```

Problem 6: Password Strength Checker

```python

def check_password(password):

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Length check
    if len(password) < 8:
        return "‚ùå Weak Password (Minimum length 8 required)"

    # Loop through password characters
    for char in password:

        if char.isdigit():
            has_digit = True

        if char in special_chars:
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "‚úÖ Strong Password"
    else:
        return "‚ùå Weak Password (Include digit & special character)"



pwd = input("\nEnter your password: ")


strength = check_password(pwd)
print(strength)
```
