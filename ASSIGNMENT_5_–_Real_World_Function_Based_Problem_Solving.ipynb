1. Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


analyze_engagement([400, 300, 350])
```

2. Healthcare – Medicine Stock Alert System

```python
def check_medicine_stock(medicine_name, stock_level):
    status = "Low Stock Alert" if stock_level < 10 else "Stock Sufficient"
    print(f"Medicine: {medicine_name} | Stock: {stock_level}")
    print(f"Status: {status}")


check_medicine_stock("Paracetamol", 6)
```

3. Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(monthly_rainfall, required_level):
    average_rf = sum(monthly_rainfall) / len(monthly_rainfall)
    status = "Adequate Rainfall" if average_rf >= required_level else "Inadequate Rainfall"

    print(f"Average Rainfall: {average_rf:.1f}")
    print(f"Rainfall Status: {status}")


check_rainfall([80, 65, 71, 72], 70)
```

4. Social Media – Duplicate Account Detection

```python
def detect_duplicates(usernames):
    # If the length of the set is smaller than the list, duplicates exist
    has_duplicates = len(set(usernames)) != len(usernames)
    result = "Yes" if has_duplicates else "No"
    print(f"Duplicate Accounts Found: {result}")


detect_duplicates(["user1", "ace_99", "user1", "dev_pro"])
```

6. Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print(f"Premium Crops: {premium_crops}")


filter_premium_crops([1500, 2500, 1800, 3200, 1900])
```

7. System Monitoring – Application Health Checker

```python
def check_app_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {errors}")
    print(f"System Status: {status}")


check_app_health(7)
```

8. Banking – Daily Transaction Limit Checker

```python
def validate_transaction(amount):
    daily_limit = 50000
    status = "Approved" if amount <= daily_limit else "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")


validate_transaction(60000)
```

9. E-Learning – Student Attendance Eligibility System

```python
def check_attendance_eligibility(attendance_record):
    days_present = 0
    for day in attendance_record:
        days_present += day

    percentage = (days_present / len(attendance_record)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print(f"Attendance Percentage: {percentage:.1f}")
    print(f"Exam Eligibility: {status}")


check_attendance_eligibility([1, 1, 1, 1, 0, 1, 1, 1, 0, 1])
```

10. Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    bill = 0

    # Calculate bill based on usage slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Determine usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


total_bill, classification = calculate_electricity_bill(250)
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Status: {classification}")
```
