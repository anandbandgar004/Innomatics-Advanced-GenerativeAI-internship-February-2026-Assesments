Problem 1 - Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert to set to get unique words
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2 - Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

Problem 3 - Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Initialize
max_val = numbers[0]
min_val = numbers[0]

# Loop comparison
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

Problem 4 - Count Products Above Price Threshold

```python
# Count Products Above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5 - Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

Problem 6 - Remove Duplicate Phone Numbers

Problem 7 - Count Character Frequency

```python
text = "python"

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

Problem 9 - Check if Key Exists in Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000
}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Problem 10 - Calculate Average Marks

```python
marks = [70, 75, 80, 65]

total = sum(marks)
count = len(marks)

average = total / count

print("Average Marks:", average)
```
