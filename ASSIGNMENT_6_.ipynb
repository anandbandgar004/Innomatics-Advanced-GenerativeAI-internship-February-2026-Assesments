# Problem 1: Smart Parking Lot Management System

```python


def smart_parking_system(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)

    if current_vehicles > capacity:
        print("Parking Status: Exceeded Capacity ðŸš¨")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)



capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

# Problem 2: Online Food Delivery Time Estimator

```python


def estimate_delivery_time(distance, traffic, weather):

    # Base time: assume 5 minutes per km
    eta = distance * 5

    # Traffic adjustment
    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 10
    elif traffic.lower() == "low":
        eta += 5

    # Weather adjustment
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "stormy":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")



estimate_delivery_time(8, "High", "Rainy")
```

# Problem 3: Movie Theatre Seat Occupancy Analyzer

```python


def seat_occupancy(total_seats, booked_seats):

    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy_percentage), "%")

    if occupancy_percentage == 100:
        print("Show Status: Housefull ðŸŽ¬")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")



total_seats = 200
booked_seats = [1] * 150

seat_occupancy(total_seats, booked_seats)
```

# Problem 4: Cloud Server Load Classification System

```python


def classify_server_load(cpu_readings):

    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(avg_cpu), "%")

    if avg_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical ðŸš¨")



cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

# Problem 5: Smart Classroom Resource Usage Monitor

```python


def resource_monitor(resource_usage):

    overused = []

    # Assume overuse threshold = 8 hours
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes âš¡")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage_data)
```

# Problem 6: Online Event Registration Capacity Controller

```python


def event_registration_controller(capacity, registrations):

    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        confirmed = capacity
        waitlisted = registrations - capacity

        print("Confirmed Registrations:", confirmed)
        print("Waitlisted Users:", waitlisted)
        print("Registration Status: Closed")


# Sample Input
event_registration_controller(100, 105)
```
