# **1. User Login Check**

```python

# TASK 1: USER LOGIN CHECK


def login_system():
    """
    This function checks whether the entered username
    and password match the predefined credentials.
    """

    # Stored credentials (Database simulation)
    stored_username = "admin"
    stored_password = "1234"

    # Taking input from user
    username = input("Enter Username: ")
    password = input("Enter Password: ")

    print("\nChecking credentials...\n")

    # Validation logic
    if username == stored_username and password == stored_password:
        print("✅ Login Successful")
    else:
        print("❌ Invalid Credentials")


# Function Call
login_system()
```

# **2. Pass / Fail Analyzer**

```python

# TASK 2: PASS / FAIL ANALYZER

def analyze_results(marks):
    """
    This function analyzes student marks and
    counts pass and fail students.
    """

    pass_count = 0
    fail_count = 0

    print("\nStudent Results Analysis\n")
    print("----------------------------")

    for index, mark in enumerate(marks, start=1):
        if mark >= 50:
            result = "PASS"
            pass_count += 1
        else:
            result = "FAIL"
            fail_count += 1

        print(f"Student {index}: Marks = {mark} --> {result}")

    print("\nSummary")
    print("----------------------------")
    print("Total Students :", len(marks))
    print("Passed Students:", pass_count)
    print("Failed Students:", fail_count)


# Sample Data
marks = [45, 78, 90, 33, 60]

# Function Call
analyze_results(marks)
```

# **3. Simple Data Cleaner**

```python

# TASK 3: SIMPLE DATA CLEANER


def clean_names(name_list):
    """
    This function removes extra spaces and
    converts names into lowercase format.
    """

    cleaned_list = []

    print("\nCleaning Names...\n")

    for name in name_list:
        print("Original Name:", repr(name))

        # Step 1: Remove extra spaces
        name_without_spaces = name.strip()

        # Step 2: Convert to lowercase
        standardized_name = name_without_spaces.lower()

        cleaned_list.append(standardized_name)

    return cleaned_list


# Sample Data
names = [" Alice ", "bob", " CHARLIE "]

# Function Call
cleaned_names = clean_names(names)

print("\nFinal Cleaned List:", cleaned_names)
```

# **4. Message Length Analyzer**

```python

# TASK 4: MESSAGE LENGTH ANALYZER


def analyze_messages(messages):
    """
    This function checks the length of each message
    and flags messages longer than 10 characters.
    """

    print("\nAnalyzing Messages...\n")
    print("---------------------------------")

    for message in messages:
        length = len(message)

        print(f"Message: '{message}'")
        print("Length :", length)

        if length > 10:
            print("⚠️  Warning: Message exceeds 10 characters.")

        print("---------------------------------")


# Sample Data
messages = ["Hi", "Welcome to the platform", "OK"]

# Function Call
analyze_messages(messages)
```

# **5. Error Message Detector**

```python

# TASK 5: ERROR MESSAGE DETECTOR


def count_errors(logs):
    """
    This function counts how many times
    'ERROR' appears in system logs.
    """

    error_count = 0

    print("\nScanning System Logs...\n")

    for index, log in enumerate(logs, start=1):
        print(f"Log Entry {index}: {log}")

        if log == "ERROR":
            error_count += 1

    print("\nTotal ERROR Entries Found:", error_count)


# Sample Data
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Function Call
count_errors(logs)
```
